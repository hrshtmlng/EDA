### Univariate Analysis

```python
df_hybrid=df.loc[df['fuelType']=='Hybrid']
```

```python
df_petrol=df.loc[df['fuelType']=='Petrol']
```

```python
plt.plot(df_hybrid['engineSize'],np.zeros_like(df_hybrid['engineSize']),'o')
plt.plot(df_petrol['engineSize'],np.zeros_like(df_petrol['engineSize']),'o')
plt.xlabel('Engine Size')
plt.show()
```

```python
df_semiAuto=df.loc[df['transmission']=='Semi-Auto']
df_automatic=df.loc[df['transmission']=='Automatic']
```

```python
plt.plot(df_semiAuto['price'],np.zeros_like(df_semiAuto['price']),'*')
plt.plot(df_automatic['price'],np.zeros_like(df_automatic['price']),'o')
plt.xlabel('Price')
```
